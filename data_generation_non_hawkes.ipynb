```python
from pathlib import Path

def random_seed(seed):
    np.random.seed(seed)
#     torch.manual_seed(seed)

def simulate_non_hawkes(n_clusters, n_nodes, n_realiz, end_time, seed, save_dir='K2_C5_non_hawkes'):
    my_inf = 100000
    T = end_time
    K = n_nodes
    baseline = np.repeat(1, K) # np.random.random(K)
    lambdas = [lambda t, k_=k: np.log(1 + np.exp(baseline[k_-1] + np.sin(t)*(1-np.cos(t+k)))) for k in range(1, K+1)]
    lambda_stars = [2.94756, 2.99087, 2.49008, 1.83082, 1.61356, 2.20026]
    tss = []
    
    for r, realiz in enumerate(range(n_realiz)):
        this_seed = seed + r
        print (r, 'realiz', this_seed)
        random_seed(this_seed)
        
        t = [0]
        i = 1
        k_k = []
        fails = 0
    
        while t[i - 1] < T:
            t_k = []
            for k in range(K):
#                 ress = [lambdas[k](t_j) for t_j in np.linspace(t[i-1], my_inf, my_inf)]
#                 lambda_star = np.max(ress)
                t_cur = t[i - 1]

                d = 0
                found = True
                while True:
                    delta = np.random.exponential(lambda_stars[k])
                    u = np.random.uniform(0, 1)
                    t_cur += delta
                    d += 1
                    if u * lambda_stars[k] <= lambdas[k](t[i-1]) or t_cur >= T:
                        if t_cur >= T:
#                             print ('fail')
                            found = False
                        else:
#                             print (d, 'cycles')
                            t_k.append(t_cur)
                        break

            if len(t_k) != 0:
                t.append(np.min(t_k))
                k_k.append(np.argmin(t_k))
                i += 1

            if found == False:
                fails += 1

            if fails > 1:
                tss.append([t[1:], k_k])
                break
            
        if fails <= 1:
            tss.append([t[1:], k_k])
            
    return tss

def simulate_clusters(n_clusters, n_nodes, n_realiz, end_time, seed=1, save_dir='K2_C5_non_hawkes'):
    save_dir = Path('', save_dir) # 'data/simulated_Hawkes',
    save_dir.mkdir(exist_ok=True, parents=True)

    clusters = []
    for i in range(n_clusters):
        print ('==============')
        print (i, 'cluster')
        seed = (i+1) * 1000
        clusters.append(simulate_non_hawkes(n_clusters, n_nodes, n_realiz, end_time, seed, save_dir=save_dir))
      
    cluster_ids = []
    dfs = []
    for cl_id, cluster in enumerate(clusters):
        cluster_ids += [cl_id] * len(cluster)
        for ts in cluster:
            df = pd.DataFrame(data=np.array(ts).T, columns=['time', 'event'])
            dfs.append(df)
    
    for i, df in enumerate(dfs):
        df.to_csv(Path(save_dir, f'{i+1}.csv').open('w'))
        
    pd.DataFrame(data=np.array(cluster_ids), columns=['cluster_id']).to_csv(Path(save_dir, f'clusters.csv').open('w'))
    return clusters
```
